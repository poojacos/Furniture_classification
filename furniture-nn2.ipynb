```python
import cv2
import numpy as np
import os
from sklearn.preprocessing import MultiLabelBinarizer
from sklearn.model_selection import train_test_split
from keras.utils import np_utils
from keras.models import Sequential
from keras.layers import Dense, Dropout, Flatten, Activation
from keras.layers import MaxPooling2D
from keras.layers import Conv2D
from PIL import Image

np.random.seed(420)
#paths
path_bed='C:/Users/Lenovo/Desktop/iadap/3-2/NN/ass2/Furniture/bed_jpg/'
path_chair='C:/Users/Lenovo/Desktop/iadap/3-2/NN/ass2/Furniture/chair_jpg/'
path_lamp='C:/Users/Lenovo/Desktop/iadap/3-2/NN/ass2/Furniture/lamp_jpg/'
path_shelf='C:/Users/Lenovo/Desktop/iadap/3-2/NN/ass2/Furniture/shelf_jpg/'
path_sofa='C:/Users/Lenovo/Desktop/iadap/3-2/NN/ass2/Furniture/sofa_jpg/'
path_stool='C:/Users/Lenovo/Desktop/iadap/3-2/NN/ass2/Furniture/stool_jpg/'
path_table='C:/Users/Lenovo/Desktop/iadap/3-2/NN/ass2/Furniture/table_jpg/'
path_wardrobe='C:/Users/Lenovo/Desktop/iadap/3-2/NN/ass2/Furniture/wardrobe_jpg/'
#image name lists
img_bed=os.listdir(path_bed)
img_chair=os.listdir(path_chair)
img_lamp=os.listdir(path_lamp)
img_shelf=os.listdir(path_shelf)
img_sofa=os.listdir(path_sofa)
img_stool=os.listdir(path_stool)
img_table=os.listdir(path_table)
img_wardrobe=os.listdir(path_wardrobe)
```

```python
images=[]
y=[]

def createDataset2(img_furniture, class_val,pathf):
    for i in img_furniture:
        if(i[:2]=='._' or i is None):
            continue
        im=cv2.imread(pathf+i,cv2.IMREAD_COLOR)
        #im=np.asarray(Image.open(pathf+i))
        im=np.asarray(cv2.resize(im, dsize=(64, 64), interpolation=cv2.INTER_CUBIC))
        images.append(im)
        y.append(class_val)

def createDataset1():
    createDataset2(img_bed,0,path_bed)  
    createDataset2(img_chair,1,path_chair)  
    createDataset2(img_lamp,2,path_lamp)  
    createDataset2(img_shelf,3,path_shelf)  
    createDataset2(img_sofa,4,path_sofa)  
    createDataset2(img_stool,5,path_stool)  
    createDataset2(img_table,6,path_table)  
    createDataset2(img_wardrobe,7,path_wardrobe)        

createDataset1()  
      
        
```

```python
    
y = np_utils.to_categorical(y, num_classes=8)
images=np.asarray(images)
y=np.array(y)

x_train,x_test,y_train,y_test=train_test_split(images,y,test_size=0.2)
x_train=x_train.reshape((len(x_train),64,64,3))
x_test=x_test.reshape((len(x_test),64,64,3))

x_train=x_train.astype(np.float32)
x_train/=255.

x_test=x_test.astype(np.float32)
x_test/=255.


```

```python
def NeuralNetwork(x_train,x_test,y_train,y_test):
    batch_size=128

    model = Sequential()
    model.add(Conv2D(64, (3, 3),input_shape=(64,64,3)))
    model.add(Activation('relu'))
    model.add(Conv2D(64, (3, 3)))
    model.add(Activation('relu'))
    model.add(MaxPooling2D(pool_size=(2, 2)))
    
    model.add(Conv2D(128, (3, 3)))
    model.add(Activation('relu'))
    model.add(Conv2D(128, (3, 3)))
    model.add(Activation('relu'))
    model.add(MaxPooling2D(pool_size=(2, 2)))
    
    model.add(Conv2D(256, (3, 3)))
    model.add(Activation('relu'))
    model.add(Conv2D(256, (3, 3)))
    model.add(Activation('relu'))
    model.add(MaxPooling2D(pool_size=(2, 2)))
    
    model.add(Flatten())
    model.add(Dense(1024))
    model.add(Activation('relu'))
    model.add(Dropout(0.25))
    model.add(Dense(1024))
    model.add(Activation('relu'))
    model.add(Dropout(0.25))
    model.add(Dense(8))
    model.add(Activation('sigmoid'))
    
    # 8. Compile model
    
    model.compile(loss='binary_crossentropy',
                  optimizer='adam',
                  metrics=['accuracy'])
    
    # 9. Fit model on training data
    model.fit(x_train, y_train, batch_size=batch_size, epochs=50,
              verbose=1, shuffle=True)

    # 10. Evaluate model on test data
    score = model.evaluate(x_test, y_test, verbose=0)
    model.save('modelserious.h5')
    print('model saved')

    print('Test score from evaluate: '+str(score))
    
NeuralNetwork(x_train,x_test,y_train,y_test)    
```
